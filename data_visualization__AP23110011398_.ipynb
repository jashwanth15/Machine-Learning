```python
cat_counts = data['Sex'].value_counts()
cat_counts
```

```python
plt.bar(cat_counts.index, cat_counts.values)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Distribution of Sex in Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(cat_counts.index, cat_counts.values)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Distribution of Sex in Titanic Dataset")
plt.xticks(rotation=45, ha='right', fontsize=12) # Rotate x-axis labels
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(cat_counts.index, cat_counts.values)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Distribution of Sex in Titanic Dataset")
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(cat_counts.index, cat_counts.values, width=0.2)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Distribution of Sex in Titanic Dataset")
plt.xticks(rotation=90, fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(cat_counts.index, cat_counts.values, width=0.2, color='orange')
plt.title('Distribution of Sex in Titanic Dataset', fontsize=15)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation = 90, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(cat_counts.index, cat_counts.values, width=0.2, color='orange')
plt.title('Distribution of Sex in Titanic Dataset', fontsize=15)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation = 90, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
sns.countplot(x='Sex', data=data, order=data['Sex'].value_counts().index, color='cornflowerblue')
plt.xticks(rotation=90)
plt.show()
```

```python
sex_counts = data['Sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Proportion of Sex in Titanic Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Age'].dropna(), bins=20, edgecolor='black') # Drop NaN values in Age column
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age in Titanic Dataset')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Age'].dropna(), bins=15, edgecolor='black') # Changed the number of bins to 15
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age in Titanic Dataset (with 15 bins)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
count, bins, _ = plt.hist(data['Age'].dropna(), bins=15, edgecolor='black') # Use 'Age' column instead of 'Year'
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age in Titanic Dataset')
plt.grid(axis='y', alpha=0.75)
plt.show()

print("Counts per bin:")
print(count)
print("\nBin edges:")
print(bins)
```

```python
sns.histplot(data=data, x='Age', bins=10) # Use 'Age' column instead of 'Year'
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age in Titanic Dataset (Seaborn)')
plt.show()
```

```python
sns.kdeplot(data['Age'].dropna()) # Use 'Age' column instead of 'Year' and drop NaN values
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution of Age in Titanic Dataset (KDE Plot)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y=data["Age"])
plt.ylabel('Age', fontsize=12)
plt.title('Distribution of Age in Titanic Dataset', fontsize=15)
plt.show()
```
